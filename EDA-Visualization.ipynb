```python
df['social_media_time_hours'] = df['social_media_time_min'] / 60
bins = [0, 18, 24, 30, 50, 70]
labels = ['0-18', '18-24', '24-30', '30-50', '50-70']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)

#  חישוב ממוצע זמן ברשתות חברתיות לפי מגדר וקבוצת גיל
usage_stats = df.groupby(['age_group', 'gender'])['social_media_time_hours'].mean().reset_index()
usage_pivot = usage_stats.pivot(index='age_group', columns='gender', values='social_media_time_hours')

# בניית הגרף
plt.figure(figsize=(10,6))
x = range(len(usage_pivot))
bar_width = 0.35

# ציור עמודות לגברים (Male)
plt.bar([i - bar_width/2 for i in x], usage_pivot['Male'], 
        width=bar_width, color='blue', label='Male')

# ציור עמודות לנשים (Female)
plt.bar([i + bar_width/2 for i in x], usage_pivot['Female'], 
        width=bar_width, color='pink', label='Female')

# עיצוב הגרף
plt.xticks(x, usage_pivot.index)
plt.ylabel("Average Social Media Time (hours)")
plt.xlabel("Age Group")
plt.title("Social Media Usage by Age Group and Gender")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.3)

plt.show()
```

```python

numerical_cols = ['age', 'daily_screen_time_min', 'social_media_time_min',
                  'negative_interactions_count', 'positive_interactions_count',
                  'sleep_hours', 'physical_activity_min', 'anxiety_level',
                  'stress_level', 'mood_level']
corr_matrix = df[numerical_cols].corr()

# יצירת גרף מפת חום 
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap: Mental Health and Social Media Usage')

#הצגת הגרף והסבר על הקשרים בין הנתונים:
#קרוב ל-1+ (צבע חם/אדום): קשר חיובי חזק. כשמשתנה אחד עולה, השני עולה איתו.
#קרוב ל-1- (צבע קר/כחול): קשר שלילי חזק. כשמשתנה אחד עולה, השני יורד
# סביב 0 אין קשר לינאר בין המספרים
plt.show()
```

```python
#פילטור להקשר בין אטרקציות חיוביות או שליליות
age_stats = df.groupby('age')[['negative_interactions_count', 'positive_interactions_count']].mean().reset_index()
#הגרף 
plt.figure(figsize=(12, 6))
sns.lineplot(data=age_stats, x='age', y='negative_interactions_count', label='Negative', color='red', marker='o')
sns.lineplot(data=age_stats, x='age', y='positive_interactions_count', label='Positive', color='green', marker='o')

plt.title('Trend: Average Interactions by Age', fontsize=16)
plt.xlabel('Age')
plt.ylabel('Average Count')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.show()
#ניתן לראות שככל שיותר צעירים יש אינטרקציות יותר חיובות מהפלטפורמות החברתיות כלומר יותר מרוצים
#באינטרקציות השלילות ניתן הבין באוכלוסיה המבוגרת יותר שפחות או יותר אין שימוש יותר גדול או יותר מועט כלומר הטווח משתנה אין לה ממש זמן ושימוש במדיה החברתית כלומר אין הפעה ניכרת על רמות מנטל סטייט
```

```python
df['date'] = pd.to_datetime(df['date'])

#יצירת רבעונים כל שנה 
df['quarter'] = df['date'].dt.to_period('Q').astype(str)

# חישוב ממוצעים רבעוניים
quarterly_stats = df.groupby('quarter')[['stress_level', 'anxiety_level', 'mood_level']].mean()

# יצירת הגרף
plt.figure(figsize=(12, 6))
plt.plot(quarterly_stats.index, quarterly_stats['stress_level'], marker='o', label='Stress Level', color='red', linewidth=2)
plt.plot(quarterly_stats.index, quarterly_stats['anxiety_level'], marker='s', label='Anxiety Level', color='green', linewidth=2)
plt.plot(quarterly_stats.index, quarterly_stats['mood_level'], marker='s', label='mood_level', color='blue', linewidth=2)

plt.title('Quarterly Mental Health Trends (2024-2025)')
plt.ylabel('Average Level (1-10)')
plt.xlabel('Quarter')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
